```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(-1,6)
plt.ylim(-1,12)
plt.title('String Title')
plt.savefig("myfirstplot.png")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.5,1,1])
axes.plot(x,y)
```
