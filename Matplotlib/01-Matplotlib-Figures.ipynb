```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
fig = plt.figure()

# LARGE AXES

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# SMALL AXES

axes2 = fig.add_axes([1,1,2,2])

axes2.plot(a,b);

axes2.set_xlim(1,2)
axes2.set_ylim(0,40)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

```python
fig = plt.figure(figsize=(3,3),dpi=200)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
fig.savefig('new_fig.png',bbox_inches='tight')
```
