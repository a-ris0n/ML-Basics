```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(10,10),dpi=100)

axes[0][0].plot(x,y);
axes[0][0].set_ylabel('Y label 0,0')

axes[0][1].plot(a,b);
axes[0][1].set_title('Title 1,0')

fig.suptitle('Figure Level Title',fontsize=16)

plt.tight_layout()
# fig.subplots_adjust(wspace=0.5,hspace=0.5)

fig.savefig('new_subplots.png', bbox_inches='tight')
```
